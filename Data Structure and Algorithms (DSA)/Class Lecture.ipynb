```python
#*******Node******

class node:
    def __init__(self, value):
        self.data = value
        self.next = None


#********Linked List*******

class linkedlist:
    def __init__(self):
        #Empty linked list
        self.head = None

        #No. of nodes in the linked list
        self.n = 0

    #Create len function
    def __len__(self):
        return self.n

    #Insert from head (Method)
    def insert_head(self, value):

        #New Node
        new_node = node(value)

        #Create connection
        new_node.next = self.head      #Location of head

        #Reassign head
        self.head = new_node           #Head get access of node attributes (self.data and self.next)

        #Increment n
        self.n += 1

    #Print Fuction -> Print linked list value
    #Change for Question no 5 solution
    def __str__(self):
        current = self.head            #Current get access of head. i.e. node attributes (self.data and self.next)
        result = ''
        while current != None:
            result = result + str(current.data)
            current = current.next
        return result

    #Insert from tail
    def append(self, value):
        new_node = node(value)

        #If LL Empty?
        if self.head == None:
            self.head = new_node
            self.n += 1
            return

        #If LL not Empty.
        current = self.head
        while current.next != None:
            current = current.next

        #You are at the last node
        current.next = new_node
        self.n += 1

    #Insert value in middle -> insert
    def insert(self, after, value):
        new_node = node(value)
        current = self.head

        #Loop through the after value
        while current != None:
            if current.data == after:
                break
            current = current.next

        #Case 1: break -> Item found and current is not None.
        if current != None:
            new_node.next = current.next
            current.next = new_node
            self.n += 1

        #Case 2: Not break -> Item not found and current is None.
        else:
            return "Item Not Found!"

    #Clear Linked List
    def clear(self):
        self.head = None
        self.n = 0

    #Delete data from head
    def delete_head(self):
        if self.head == None:
            return "Empty Linked List"
            
        self.head = self.head.next
        self.n -= 1

    #Delete form tail -> pop
    def pop(self):
        current = self.head

        #If Linked list have only one node/item
        if current.next == None:
            return self.delete_head()

        while current.next.next != None:
            current = current.next
        current.next = None
        self.n -= 1

    #Delete by Value -> remove
    def remove(self, value):

        #Check the Linked list is empty or not!
        if self.head == None:
            return "Empty Linked List"

        #If remove the firest element
        if self.head.data == value:
            return delete_head()

        current = self.head
        while current.next != None:
            if current.next.data == value:
                break
            current = current.next

        #case 1: If value not found
        if current.next == None:
            return "value not Found!"

        #case 2: If value found
        else:
            current.next = current.next.next

    #Search by Value -> Find
    def find(self, value):
        current = self.head
        position = 0

        while current != None:
            if current.data == value:
                return position
            current = current.next
            position += 1
        return "Not Found"

    #Search by index
    def __getitem__(self, index):
        current = self.head
        position = 0
        while current != None:
            if position == index:
                return current.data
            current = current.next
            position += 1
        return "IndexError"
        
    # Question no 1 solution
    def function(self):
        current = self.head
        if current == None:
            return
        if current.next.next != None:
            print(current.data, ' ', end = '')
            function(current.next)
        print(current.data, '', end = '')

    #Queation no 2 solution
    def replace_max(self, value):
        current = self.head
        maximum = current

        while current != None:
            if current.data > maximum.data:
                maximum = current
            current = current.next
        
        maximum.data = value

    #Question no 3 solution
    def sum_odd_position(self):
        current = self.head
        position = 0
        summation = 0

        while current != None:
            if position % 2 != 0:
                summation += current.data
            current = current.next
            position += 1
        return summation

    #Question no 4 solution
    def reverse(self):
        prev_node = None
        current = self.head

        while current != None:
            next_node = current.next
            current.next = prev_node
            prev_node = current
            current = next_node
            
        self.head = prev_node

    #Question no 5 solution
    def string(self):
        current = self.head

        while current != None:
            if current.data == '*' or current.data == '/':
                current.data = ' '
                if current.next.data == '*' or current.next.data == '/':
                    current.next.next.data = current.next.next.data.upper()
                    current.next = current.next.next
            current = current.next
```

## Stack

**Create Stack by using Linked List**


**Create Node**

```python
class node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

**Create Stack**

```python
class stack:
    def __init__(self):
        self.top = None

    #Check the stack is empty!
    def isempty(self):
        return self.top == None

    #Input value -> push
    def push(self, value):
        new_top = node(value)
        new_top.next = self.top
        self.top = new_top

    #See top value -> peak
    def peak(self):
        if(self.isempty()):
            return 'Empty Stack'
        else:
            return self.top.value

    #Delete top value -> pop
    def pop(self):
        if(self.isempty()):
            return 'Empty Stack'
        else:
            data = self.top.value
            self.top = self.top.next
            return data

    #Size of Stack
    def size(self):
        current = self.top
        count = 0

        while current != None:
            count += 1
            current = current.next
        return count

    #Traverse
    def traverse(self):
        current = self.top

        while current != None:
            print(current.value, end= ' ')
            current = current.next
```

1. Write a Python Program to reverse string by using stack.

Input: `Hello`

Output: `olleH`

```python
def reverse_string(text):
    s = stack()

    for i in text:
        s.push(i)

    result = " "
    while(not s.isempty()):
        result = result + s.pop()
    print(result)
```

```python
reverse_string("Hello")
```

2. Write a Python program to solve the problem below.

- Input: `Hello`
- Pattern: `uuur`
- Output: `Hel`

Where `u`menas `Undo` and `r` means `redo`.

```python
def undo_redo(text, pattern):
    u = stack()
    r = stack()

    for i in text:
        u.push(i)

    for i in pattern:
        if i == "u":
            undo = u.pop()
            r.push(undo)
        else:
            redo = r.pop()
            u.push(redo)
    
    result = ""
    while(not u.isempty()):
        result = u.pop() + result
    return result
```

```python
undo_redo("Hello", "uuur")
```

3. Celebrity Problem.

Input: Input a `n*n` matrix where `0` means don't know and `1` means know.  
Ouput: Find celebrity who don't know anyone but everyone know his/him.

```python
def celebrity(matrix):
    s = stack()

    for i in range(len(matrix)):
        s.push(i)

    while s.size() >= 2:
        i = s.pop()
        j = s.pop()

        if matrix[i][j] == 0:
            # i don't know j. -> j is not celebrity
            s.push(i)
        else:
            # j don't know i. -> i is not celebrity
            s.push(j)

    celeb = s.pop()
    for i in range(len(matrix)):
        if i != celeb:
            if matrix[i][celeb] == 0 or matrix[celeb][i] == 1:
                return "No one is Celebrity"
    print("The Celebrity is", celeb)
```

**Python list as fixed Array**

```python
class stack:
    def __init__(self, size):
        self.size = size
        self.stack = [None] * self.size
        self.top = -1

    def push(self, value):
        if self.top == self.size - 1:
            return "Overflow"
        else:
            self.top += 1
            self.stack[self.top] = value
    
    def pop(self):
        if self.top == -1:
            return "Empty"
        else:
            data = self.stack[self.top]
            self.top -= 1
            return data

    def traverse(self):
        for i in range(self.top + 1):
            print(self.stack[i], end = " ")
```

## Queue

**Create Node**

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None
```

**Create Queue by using Linked List**

```python
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, value):
        new_node = node(value)

        if self.front == None:
            self.front = new_node
            self.rear = self.front
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.front == None:
            return "Empty"
        else:
            data = self.front.data
            self.front = self.front.next
            return data

    def front_item(self):
        if self.front == None:
            return "Empty"
        else:
            return self.front.data

    def rear_item(self):
        if self.front == None:
            return "Empty"
        else:
            return self.rear.data

    def isempty(self):
        return self.front == None

    def size(self):
        current = self.front
        count = 0

        while current != None:
            count += 1
            current = current.next
        return count

    def traverse(self):
        current = self.front

        while current != None:
            print(current.data, end = " ")
            current = current.next
```

## Linear Search

```python
def linear_search(array, item):
    for i in range(len(array)):
        if array[i] == item:
            return i
    return -1
```

**While Loop**

```python
def binarry_search(array, item):
    low = 0
    high = len(array) - 1

    while low <= high:
        mid = (low + high) // 2

        #Check if item is present at mid
        if array[mid] == item:
             return mid
        
        #If item is grater than mid -> ignore left half
        elif array[mid] < item:
            low = mid + 1

        #If item is less than mid -> ignore right half
        else:
            high = mid - 1
    
    return -1
```

## Sorting

- Write a Python code to justify the given array is sorted or not?

```python
def is_sorted(array):
    sort = True

    for i in range(len(array)-1):
        if array[i] > array[i+1]:
            sort = False
    return sort
```

### Monkey Sort

- Sorted an array by using random shuffle technique. It doing suffle until the array get sorted.

```python
import random

def monkey_sort(array):
    while not is_sorted(array):
        random.shuffle(array)
    return array
```

```python
monkey_sort([1,2,4,3, 5])
```

### Bubble Sort

```python
def bubble_sort(array):

    for i in range(len(array)-1):
        for j in range(len(array)-1-i):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

```python
bubble_sort([1,3,2,4,6,5])
```

*Make Bubble Sort Algorithm to Adaptive Sort*

```python
def bubble_sort(array):
    for i in range(len(array)-1):
        flag = 0
        for j in range(len(array)-1-i):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
            flag = 1
        if flag == 0:
            break
    return array
```

### Selection Sort

```python
def selection_sort(array):

    for i in range(len(array)-1):
        minimum = i
        for j in range(i+1, len(array)):
            if array[j] < array[minimum]:
                minimum = j
        array[i], array[minimum] = array[minimum], array[i]
    return array
```

```python
selection_sort([1,2,4,3,7,6,5])
```

### Merge Sort

```python
def merge_sorted(arr1, arr2):

    i = j = 0
    merged = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1

    while i < len(arr1):
        merged.append(arr1[i])
        i += 1
        
    while j < len(arr2):
        merged.append(arr2[j])
        j += 1
        
    return merged
```

```python
merge_sorted([1,3,5], [2,4,6])
```

```python
def merge_sort(array):
    if len(array) == 1:
        return array

    mid = len(array) // 2

    left = array[:mid]
    right = array[mid:]

    left = merge_sort(left)
    right = merge_sort(right)
    
    return merge_sorted(left, right)
```
