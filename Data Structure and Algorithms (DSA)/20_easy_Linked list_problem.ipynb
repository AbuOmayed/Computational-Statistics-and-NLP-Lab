1. **Create a Linked List**

- Write a program to create a singly linked list and display its elements.

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_value

    def display(self):

        current = self.head
        while current:
            print(current.data, end = " -> ")
            current = current.next
        print("None")
```

```python
l = linked_list()
l.append(5)
l.append(6)
l.append(7)
```

```python
l.display()
```

`Example`

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def insert(self, value):
        new_value = node(value)
        new_value.next = self.head
        self.head = new_value

    def insert_at(self, position, value):
        new_value = node(value)

        if position == 1:
            self.head = new_value
            return

        current = self.head
        count = 1
        while current:
            count += 1
            if count == position:
                break
            current = current.next
                
        if current != None:
            new_value.next = current.next
            current.next = new_value
        else:
            return "Position is not found"

    def display(self):
        current = self.head

        while current:
            print(current.data, end = " -> ")
            current = current.next
        print("None")
```

11. Reverse a Linked List

Reverse the linked list.

`Example` 
- Input: 10 → 20 → 30 → 40
- Output: 40 → 30 → 20 → 10

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def insert(self, value):
        new_value = node(value)
        new_value.next = self.head
        self.head = new_value

    def reverse(self):

        prev_node = None
        current = self.head

        while current:
            next_node = current.next
            current.next = prev_node
            prev_node = current
            current = next_node
            
        self.head = prev_node

    def __str__(self):
        current = self.head
        result = ""
        while current:
            result = result + str(current.data) + " -> "
            current = current.next
        return result[:-4]
```

12. Find Middle Node

Find the middle element of the linked list.

`Example`
- Input: 10 → 20 → 30 → 40 → 50
- Output: 30

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)
        if self.head is None:
            self.head = new_value
            return
        current = self.head
        while current.next:
            current = current.next

        current.next = new_value

    def middle_value(self):
        if self.head is None:
            return "Empty Linked List"
            
        current = self.head
        count = 0
        while current:
            count += 1
            current = current.next
            
        mid_value = (count + 1)//2

        temp = self.head
        count_1 = 0
        while temp:
            count_1 += 1
            if count_1 == mid_value:
                return temp.data
            temp = temp.next

    def __str__(self):
        current = self.head
        result = ""

        while current:
            result = result + str(current.data) + " -> "
            current = current.next
        return result[:-4]
```

13. Remove Duplicates

Remove duplicate elements from a sorted linked list.

`Example`  
- Input: 10 → 10 → 20 → 20 → 30
- Output: 10 → 20 → 30

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_value

    def remove_duplicates(self):

        current = self.head
        
        while current and current.next:
            if current.data == current.next.data:
                current.next = current.next.next
            else:
                current = current.next

    def __str__(self):

        if self.head is None:
            return "Empty Linked List"

        current = self.head
        result = ""
        while current:
            result = result + str(current.data) + " -> "
            current = current.next
        return result[:-4]
```

Remove duplicate elements from a unsorted linked list.

- Input: 10 → 10 → 20 → 10 → 20 → 30

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_value

    def remove_duplicates(self):

        current = self.head
        prev = None
        unique = set()

        while current:
            if current.data in unique:
                prev.next = current.next
            else:
                unique.add(current.data)
                prev = current
            current = current.next

    def __str__(self):

        if self.head is None:
            return "Empty Linked List"

        current = self.head
        result = ""
        while current:
            result = result + str(current.data) + " -> "
            current = current.next
        return result[:-4]
```

16. Get Nth Node from End

Find the Nth node from the end.

`Example`
- Input: 10 → 20 → 30 → 40 → 50
- N = 2
- Output: 40

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_value

    def search(self, n):
        first = self.head
        second = self.head

        for _ in range(n):
            if first is None:
                return "Invalid N"
            first = first.next

        while first:
            first = first.next
            second = second.next
        return second.data
```

17. Sum of Nodes

Find the sum of all nodes.

`Example`
- Input: 10 → 20 → 30
- Output: 60

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_value

    def sum_ll(self):
        if self.head is None:
            return 0
            
        summation = 0
        current = self.head
        while current:
            summation += current.data
            current = current.next
        return summation
```

19. Find Maximum Element

Find the largest element in the linked list.

`Example`
- Input: 10 → 50 → 20 → 30
- Output: 50

```python
class node:
    def __init__(self, value):
        self.data = value
        self.next = None

class linked_list:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_value = node(value)

        if self.head is None:
            self.head = new_value
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_value

    def max_value(self):
        if self.head is None:
            return "Empty Linked List"

        current = self.head
        first = self.head

        while current:
            if first.data < current.data:
                first = current
            current = current.next
        return first.data
```
