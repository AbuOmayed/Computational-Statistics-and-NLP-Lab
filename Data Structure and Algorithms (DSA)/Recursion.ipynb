1. Factorial of a Number

Compute factorial using recursion.  
`Example:` 5! = 120

```python
def factorial(n):
    if n <= 1:
        return n
    return n * factorial(n-1)
```

```python
factorial(5)
```

2. Fibonacci Number

Find the nth Fibonacci number.  
`Example:` fib(5) = 5

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5)
```

3. Sum of Natural Numbers

Find sum from 1 to n.  
`Example`: n=5 → 15

```python
def summation(n):
    if n <= 1:
        return n
    return n + summation(n-1)
```

```python
summation(5)
```

4. Print Numbers from 1 to N

Print numbers recursively.

```python
def print_number(n):
    if n == 0:
        return
        
    print_number(n-1)

    print(n, end = " ")
```

```python
print_number(5)
```

5. Print Numbers from N to 1

Reverse order printing.

```python
def print_number(n):
    if n == 0:
        return

    print(n, end = " ")

    print_number(n-1)
```

```python
print_number(5)
```

6. Sum of Digits

Find sum of digits of a number.  
`Example:` 123 → 6

```python
def sum_digit(n):
    if n == 0:
        return n
    return n % 10 + sum_digit(n//10)
```

```python
sum_digit(1234)
```

7. Reverse a String

Reverse a string using recursion.  
`Example:` "abc" → "cba"

```python
def reverse(s):
    if len(s) == 1:
        return s
    return reverse(s[1:]) + s[0]
```

```python
reverse("abcd")
```

8. Check Palindrome (String)

Check if a string is palindrome.

```python
def palindrom(s, left, right):
    if left >= right:
        return "Palindorm"
    if s[left] != s[right]:
        return "Not Palindrim"
    return palindrom(s, left+1, right-1)

def palindrom_check(s):
    left = 0
    right = len(s) - 1
    return palindrom(s, left, right)
```

```python
palindrom_check("anna")
```

9. Power Function

Compute $x^{n}$ recursively.

```python
def power(x,n):
    if n == 1:
        return x
    return x*power(x, n-1)
```

```python
power(8,2)
```

10. Count Digits

Count number of digits in an integer.

```python
def count_digit(s):
    if s == 0:
        return 0
    return 1 + count_digit(s//10)
```

```python
count_digit(1234)
```

11. Find Maximum in a List

Find max element using recursion.

```python
def maximum(l, left, right, maxi):
    if left >= right:
        return maxi
    if l[left] > maxi:
        maxi = l[left]
    if l[right] > maxi:
        maxi = l[right]
    return maximum(l, left+1, right-1, maxi)

def find_max(l):
    maxi = l[0]
    left = 0
    right = len(l) - 1
    return maximum(l, left, right, maxi)
```

```python
find_max([4, 6, 7, 1, 2, 3, 90, 2, 4])
```

12. Find Minimum in a List

Find min element recursively.

```python
def minimum(l, left, right, mini):
    if left >= right:
        return mini
    if l[left] < mini:
        mini = l[left]
    if l[right] < mini:
        mini = l[right]
    return minimum(l, left+1, right-1, mini)

def find_min(l):
    mini = l[0]
    left = 0
    right = len(l) - 1
    return minimum(l, left, right, mini)
```

```python
find_min([30, 40 ,20, 50])
```

13. Linear Search (Recursive)

Search an element in a list.

```python
def search(l, x, left, right):
    if left > right:
        return "Not Found"
    if l[left] == x:
        return left
    if l[right] == x:
        return right
    return search(l, x, left + 1, right - 1)
    
def linear_search(l, x):
    left = 0
    right = len(l) - 1
    return search(l, x, left, right)
```

```python
linear_search([1,2,3,4], 3)
```

`Alternative`

```python
def search(arr,x, i=0):
    if i > len(arr) -1:
        return "Not Found"
    if arr[i] == x:
        return i
    return search(arr, x, i+1)
```

```python
search([1,2,3,4], 5)
```

14. Binary Search (Recursive)

Perform binary search using recursion.

```python
def search(arr, x, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == x:
        return mid
    elif arr[mid] > x:
        right = mid - 1
    else:
        left = mid + 1
    return search(arr, x, left, right)

def binarry_search(arr, x):
    left = 0
    right = len(arr) - 1
    return search(arr, x, left, right)
```

```python
binarry_search([1,2,3,4,5,6,7], 8)
```

`Alternative`

```python
def search(arr, x, left, right):
    if left <= right:
        mid = (left + right) // 2

        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return search(arr, x, left, mid-1)
        else:
            return search(arr, x, mid + 1, right)
    return -1

def binarry_search(arr, x):
    left = 0
    right = len(arr) - 1
    return search(arr, x, left, right)
```

```python
binarry_search([1,2,3,4,5,6,7], 7)
```

15. GCD of Two Numbers

Use recursion (Euclidean algorithm).

```python
def GCD(a,b):
    if b == 0:
        return a
    return GCD(b , a%b)
```

```python
GCD(48,18)
```
