1. Maximum Sum Subarray of Size K

Find the maximum sum of a subarray of size K.

Example: `[2,1,5,1,3,2], K=3 → 9`

```python
def maxsum_subarray(arr, k):
    sum_subarray = sum(arr[:k])
    max_sum = sum_subarray
    left = 0

    for right in range(k, len(arr)):
        sum_subarray += arr[right]
        sum_subarray -= arr[left]
        left += 1
        max_sum = max(max_sum, sum_subarray)
    return max_sum
```

```python
maxsum_subarray([2,1,5,1,3,2], 3)
```

2. Minimum Sum Subarray of Size K

Find the minimum sum of any subarray of size K.

```python
def minsum_subarray(arr, k):
    sum_subarray = sum(arr[:k])
    minsum = sum_subarray
    left = 0

    for right in range(k, len(arr)):
        sum_subarray += arr[right]
        sum_subarray -= arr[left]
        left += 1
        minsum = min(minsum, sum_subarray)
    return minsum
```

```python
minsum_subarray([2,1,5,1,3,2], 3)
```

3. Average of Subarrays of Size K

Compute the average for all subarrays of size K.

```python
def average_subarray(arr, k):
    summation = sum(arr[:k])
    left = 0
    average_list = []

    for right in range(k, len(arr)):
        average_list.append(round(summation/k, 2))
        summation += arr[right]
        summation -= arr[left]
        left += 1

    average_list.append(round(summation/k, 2))
    
    return average_list        
```

```python
average_subarray([2,1,5,1,3,2], 3)
```

`Alternative`

```python
def average_subarray(arr, k):
    summation = sum(arr[:k])
    left = 0
    average_list = []

    for right in range(k, len(arr)+1):
        average_list.append(round(summation/k, 2))

        if right < len(arr):
            summation += arr[right]
            summation -= arr[left]
            left += 1
    
    return average_list
```

```python
average_subarray([2,1,5,1,3,2], 3)
```

`Alternative`

```python
def average_subarray(arr, k):
    left = 0
    average_list = []
    window_sum = 0

    for right in range(len(arr)):
        window_sum += arr[right]

        if right >= k-1:
            average_list.append(round(window_sum/k, 2))
            window_sum -= arr[left]
            left += 1
    
    return average_list
```

```python
average_subarray([2,1,5,1,3,2], 3)
```

4. Maximum Element in Each Window

Find the maximum element in each window of size K.

```python
def max_element_each_window(arr, k):
    window = arr[:k]
    result = []

    for right in range(k, len(arr)+1):
        result.append(max(window))

        if right < len(arr):
            window.append(arr[right])
            window.pop(0)
            
    return result
```

```python
max_element_each_window([2,1,5,1,3,2], 3)
```

`Alternative`

```python
def max_element_each_window(arr, k):
    window = arr[:k]
    result = []

    for right in range(k, len(arr)):
        result.append(max(window))
        window.append(arr[right])
        window.pop(0)

    result.append(max(window))
            
    return result
```

```python
max_element_each_window([2,1,5,1,3,2], 3)
```

5. Count Subarrays of Size K with Sum > X

Count how many subarrays of size K have sum greater than X.

```python
def sum_more_than_x(arr, k, x):
    summation = sum(arr[:k])
    result = []
    count = 0

    for right in range(k, len(arr)+1):
        if summation > x:
            result.append(summation)
            count += 1

        if right < len(arr):
            summation += arr[right]
            summation -= arr[right - k]

    return result, count
```

```python
sum_more_than_x([2,1,5,1,3,2], 3, 7)
```
