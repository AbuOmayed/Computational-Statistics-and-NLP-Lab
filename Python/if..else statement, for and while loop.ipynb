## While Loop

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

```python
i = 1
while i < 6:
    if i == 3:
        break
    print(i)
    i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

## For Loop

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
for x in "banana":
  print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```python
for x in range(2, 10, 2):
  print(x)
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

```python
for x in [0, 1, 2]:
  pass
```

```python
fruits = ["apple", "banana", "cherry"]

for i, j in enumerate(fruits):
    print(i, j)
```
