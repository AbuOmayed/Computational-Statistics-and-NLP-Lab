### Sets Logic Operations

- Remember that with sets you can check the difference between sets, as well as the symmetric difference, intersection and union.

```python
# Consider the following two set
A = {1,2,3,4,5}
B = {3,4,6,8,9}
```

```python
print(A,B)
```

- You can find the intersection of two sets as follow using & or intersection() method.

```python
intersection = A & B
```

```python
intersection
```

```python
intersection_1 = A.intersection(B)
```

```python
intersection_1
```

- You can find all the elements that are only contain in A using the difference() method.

```python
difference = A.difference(B)
```

```python
difference
```

- The union corresponds to all the elements in both sets, we can find union by using union() mehtod.

```python
union = A.union(B)
```

```python
union
```
