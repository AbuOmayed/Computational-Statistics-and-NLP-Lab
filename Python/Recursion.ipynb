1. Factorial of a Number

Compute factorial using recursion.  
`Example:` 5! = 120

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

2. Fibonacci Number

Find the nth Fibonacci number.  
`Example:` fib(5) = 5

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```

```python
fib(5)
```

3. Sum of Natural Numbers

Find sum from 1 to n.  
`Example`: n=5 → 15

```python
def summation(n):
    if n<=1:
        return n
    return n + summation(n-1)
```

```python
summation(5)
```

4. Print Numbers from 1 to N

Print numbers recursively.

```python
def number(n):
    if n == 0:
        return
    number(n-1)
    print(n, end = " ")
```

```python
number(10)
```

5. Print Numbers from N to 1

Reverse order printing.

```python
def number(n):
    if n == 0:
        return
    print(n, end = " ")
    number(n-1)
```

```python
number(10)
```

6. Sum of Digits

Find sum of digits of a number.  
`Example:` 123 → 6

```python
def sum_digit(n):
    if n <= 1:
        return n
    return n % 10 + sum_digit(n // 10)
```

```python
sum_digit(123)
```

7. Reverse a String

Reverse a string using recursion.  
`Example:` "abc" → "cba"

```python
def reverse_string(s):
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]
```

```python
reverse_string("abcd")
```

8. Check Palindrome (String)

Check if a string is palindrome.

```python
def palindrom_check(s, left, right):
    if left >= right:
        return 1
    
    if s[left] != s[right]:
        return 0
    
    return palindrom_check(s, left+1, right-1)

def is_palindrom(s):
    left = 0
    right = len(s) - 1
    return palindrom_check(s, left, right)
```

```python
print(is_palindrom("1aba1"))
```

9. Power Function

Compute $x^{n}$ recursively.

```python
def power(x,n):
    if n <= 1:
        return x
    return x*power(x,n-1)
```

```python
power(2,5)
```

`Alternative`

```python
def power(x, n):
    if n <= 1:
        return x
    half = power(x, n//2)

    if n % 2 == 0:
        return half*half
    else:
        return x*half*half
```

```python
power(2,4)
```

11. Find Maximum in a List

Find max element using recursion.

```python
def maximum(l, max_value, left, right):
    if left > right:
        return max_value
    if l[left] > max_value:
        max_value = l[left]
    if l[right] > max_value:
        max_value = l[right]
    return maximum(l, max_value, left+1, right-1)

def maximum_value(l):
    left = 0
    right = len(l) - 1
    max_value = l[left]
    return maximum(l, max_value, left, right)
```

```python
maximum_value([10, 25, 5, 40, 15])
```

`Alternative`

```python
def find_max(arr, n):
    # Base case
    if n == 1:
        return arr[0]

    # Recursive case
    return max(arr[n-1], find_max(arr, n-1))


# Example
arr = [10, 25, 5, 40, 15]
print("Maximum:", find_max(arr, len(arr)))
```

`Alternative`

```python
def find_max(arr):
    if len(arr) == 1:
        return arr[0]

    return max(arr[0], find_max(arr[1:]))


print(find_max([10, 25, 5, 40, 15]))
```

`Alternative`

```python
def find_max_optimized(arr, index=0):
    if index == len(arr) - 1:
        return arr[index]
    return max(arr[index], find_max_optimized(arr, index + 1))
```

```python
find_max_optimized([10, 25, 5, 40, 15])
```

12. Find Minimum in a List

Find min element recursively.

```python
def min_value(l, min_val, left, right):
    if left > right:
        return min_val
    if l[left] < min_val:
        min_val = l[left]
    if l[right] < min_val:
        min_val = l[right]
    return min_value(l, min_val, left + 1, right - 1)

def minimum(l):
    left = 0
    right = len(l) - 1
    min_val = l[left]
    return min_value(l, min_val, left, right)
```

```python
arr = [4,2,5,10]
min_val = arr[0]
min_value(arr, min_val, 0, len(arr)-1)
```

```python
minimum([4,2,5,10])
```

`Alternative`

```python
def minimum(arr):
    if len(arr) == 1:
        return arr[0]
    return min(arr[0], minimum(arr[1:]))
```

```python
minimum([4,2,5,10])
```

13. Linear Search (Recursive)

Search an element in a list.

```python
def search(arr, x):
    l = len(arr)
    for i in range(l):
        if arr[i] == x:
            return i
    return -1
```

```python
search([4,2,5,10], 5)
```

```python
def search(arr, x, i = 0):
    if arr[i] == x:
        return i
    if i == len(arr)-1:
        return -1
    return search(arr, x, i+1)
```

```python
search([4,2,5,10], 11)
```

14. Binary Search (Recursive)

Perform binary search using recursion.

```python
def search(arr, x, low, high):

    if low <= high:
        mid = (low + high)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    else:
        return -1
        
    return search(arr, x, low, high)
```

```python
arr = [1,2,3,4,5,6,7,8]
search(arr, 6, 0, len(arr)-1)
```

```python
def search(arr, x, low, high):

    if low <= high:
        mid = (low + high)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    else:
        return -1
    return search(arr, x, low, high)

def binarry_search(arr, x):
    low = 0
    high = len(arr) - 1
    return search(arr, x, low, high)
```

15. GCD of Two Numbers

Use recursion (Euclidean algorithm).

```python
def GCD(a,b):
    if b == 0:
        return a
    return GCD(b, a%b)
```

```python
GCD(18, 48)
```
