### Tuple
- Tuples are an ordered sequence.
- Tuples are written as comma separated elements within parenthesis.
- Tuples are immutable in nature i.e. the elements in the tuple cannot be added or removed once created.

```python
tuple_1 = (1,2,3,4)
```

```python
type(tuple_1)
```

```python
tuple_1[0] = 5 # Tuples are immutable
```

```python
tuple_1.append(5) # Tuples are immutable
```

```python
del(tuple_1[0]) # Tuples are immutable
```

- We can combined the Tuple by adding.

```python
A = (1,2,3)
B = (4,5,6)
C = A+B
```

```python
C
```

### Tuple method
- <b>count()</b>	Returns the number of times a specified value occurs in a tuple
- <b>index()</b>	Searches the tuple for a specified value and returns the position of where it was found

- count() Returns the number of times a specified value occurs in a tuple
- syntax: tuple.count(value)

```python
data_3 = (10, 20 ,30, 10, 40, 20, 10)
```

```python
data_3.count(10)
```

- index() Searches the tuple for a specified value and returns the position of where it was found
- tuple.index(value)

```python
data_3.index(10)
```

```python
data_3.index(40)
```
