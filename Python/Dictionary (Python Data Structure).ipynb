### Dictionary
- To create a dictionary, we use curly {} brackets.
- The key have to be immutable and unique.
- Each key and value pair is separated by comma.

```python
dic = {"Thriller": 1982, "Black to Black": 1980}
```

```python
dic
```

```python
type(dic)
```

- The key is used to look at the value.

```python
dic["Thriller"]
```

- We can add a new entry to the dictionary.

```python
dic["Graduation"] = 2007
```

```python
dic
```

- We can delete an entry.

```python
del(dic["Graduation"])
```

```python
dic
```

- We can verify if an element is in the dictionary using the "in" command.

```python
"Thriller" in dic
```

- In order to see all the keys in the dictionary.

```python
dic.keys()
```

- In the same way we can obtain values.

```python
dic.values()
```

- Dictionary of dictionary

```python
europe = {"Spain": {"Capital": "Madrid", "Population": 46.77},
         "Franch": {"Capital": "Paris", "Population": 66.03},
         "Germany": {"Capital": "Berline", "Population": 80.62}}
```

```python
europe
```

- Print out capital of Franch.

```python
europe["Franch"]["Capital"]
```

- Add details about Italy

```python
europe["Italy"] = {"Capital": "Rome", "Population": 59.83}
```

```python
europe
```
