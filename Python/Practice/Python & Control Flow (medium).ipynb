```python
def reverse(s):
    reverse_string = ""
    for i in s:
        reverse_string = i + reverse_string
    return reverse_string
```

**3** `String Compression`: Write a program that compresses a string like `"aaabbcddd"` into `"a3b2c1d3"`. If the compressed string is not smaller than the original, return the original.

`Correct Method`

```python
def anagrams(s,a):
    dic_s = {}
    dic_a = {}

    for i,j in zip(s,a):
        if i not in dic_s:
            dic_s[i] = 1
        else:
            dic_s[i] += 1
        if j not in dic_a:
            dic_a[j] = 1
        else:
            dic_a[j] += 1
    
    if len(s) != len(a):
        print("Not Anagrams")
    elif dic_s == dic_a:
        print("Anagrams")
    else:
        print("Not Anagrams")
```

**6** `The "Title Case" Exception`: Write a loop that converts a sentence to Title Case (capitalizing the first letter of every word), but excludes specific articles like "and", "the", "of" (unless they are the first word of the sentence).

**Input:** `"war and peace"`

**Output:** `"War and Peace"`

```python
def titlecase(strs):
    custom_stopwords = ["a", "an", "the", "is", "in", "and", "but", "or", "of", "on", "with"]
    new_list = strs.strip(".,!#").split()
    title_list = []
    for j in new_list:
        if j not in custom_stopwords:
            title_list.append(j.capitalize())
        else:
            title_list.append(j)
    return " ".join(title_list)
```

**8** `Rotate Array`: Write a script that rotates a list to the right by k steps.

**Input:** [1, 2, 3, 4, 5], k=2 

**Output:** [4, 5, 1, 2, 3].

```python
def rotationarray(num_list, k = 2):
    for i in range(0,k):
        temporal = num_list[len(num_list) -1]
        for j in range(len(num_list) - 1, 0, -1):
            num_list[j] = num_list[j-1]
        num_list[0] = temporal
    return num_list
```

```python
rotationarray([1, 2, 3, 4, 5])
```

**9** `Running Sum (Cumulative)`: Write a loop that transforms a list of numbers into a cumulative sum list.

**Input:** `[1, 2, 3, 4]` 

**Output:** `[1, 3, 6, 10]`

```python
def cumsum(num_list):
    cum_list = []
    total = 0
    for i in num_list:
        total += i
        cum_list.append(total)
    return cum_list
```

**10** `Merge Sorted Lists`: Without using `.sort()`, write a while loop to merge two already sorted lists `[1, 4, 6]` and `[2, 3, 5]` into a single sorted list `[1, 2, 3, 4, 5, 6]`.

[Bubble Sort](https://www.geeksforgeeks.org/dsa/bubble-sort-algorithm/)

```python
def sortlist(list_1, list_2):
    marge = list_1 + list_2
    n = len(marge)
    
    for i in range(n-1):
        for j in range(n-i-1):
            if marge[j] > marge[j+1]:
                marge[j], marge[j+1] = marge[j+1], marge[j]
            
    return marge
```

```python
sortlist([1,4,6], [2,3,5])
```

`Alternative`

```python
def sortlist(list_1, list_2):
    add = list_1 + list_2
    marge = []

    while add:
        smallest = add[0]
        for num in add:
            if num < smallest:
                smallest = num
        marge.append(smallest)
        add.remove(smallest)
    return marge
```

```python
sortlist([1,4,6], [2,3,5])
```

**13** `Grid Search`: Given a 3x3 grid (list of lists) filled with 0s and 1s, count how many 1s are touching the edges of the grid.

`n*n Grid`

```python
def gridcount(grid):
    count = 0
    n = len(grid)
    m = len(grid[0])
    for i in range(n):
        for j in range(m):
            if i == 0 or i == n-1 or j == 0 or j == m-1:
                if grid[i][j] == 1:
                    count += 1
    return count
```

**14** `Invert a Dictionary`: Given a dictionary `{ 'A': 10, 'B': 20, 'C': 10 }`, create a new dictionary where the values become the keys. Since values repeat, the new values should be lists.

**Output:** `{ 10: ['A', 'C'], 20: ['B'] }`

```python
def dictodic(dic):
    new_dic = {}
    for key, value in dic.items():
        if value not in new_dic:
            new_dic[value] = [key]
        elif value in new_dic:
            new_dic[value] += key
    return new_dic
```

**Input:** `"{[()]}"` $\rightarrow$ `True`

**Input:** `"{[(])}"` $\rightarrow$ `False`

```python
def parentheses(strs):
    list_a = [i for i in strs]
    valid = True
    
    for i in range(len(list_a)):
        if list_a[i] == list_a[- (i+1)]:
            valid = True
        else:
            valid = False
            break
            
    return valid
```
