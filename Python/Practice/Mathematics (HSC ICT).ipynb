- If a student attends five exams and achieves an average of 40 marks, they pass the exam; otherwise, they fail the exam.

```python
def student_result():
    name = input("Enter your name: ")
    n = int(input("Enter your total subjects."))
    marks = []
    for i in range(n):
        m = float(input(f"Enter marks of subject {i+1}: "))
        marks.append(m)
    total_marks = 0
    for i in marks:
        total_marks += i
    average = total_marks / len(marks)
    if average < 40:
        result = "Fail"
    elif average >= 40:
        result = "Pass"
    else: result = "Fail"
    print(f"{name} you {result} in the exam.")
```

- Write a program to calculate the **n-th term** and the **sum of the first n terms** of an **Arithmetic Progression (A.P.)**.

**Given Formulas:**
- **n-th term**:
  t_n = $a + (n - 1)d$

- **Sum of first n terms**:
  S_n = $\frac{n}{2}\,[2a + (n - 1)d]$

**Where:**
- `a` = first term  
- `d` = common difference  
- `n` = number of terms  

**Instructions:**
- Take the values of `a`, `d`, and `n` as input.
- Calculate the n-th term of the A.P.
- Calculate the sum of the first n terms.
- Display the calculated results.


```python
def arithmetic_progression():
    a = float(input("Enter first term of the Arithmetic Progression: "))
    d = float(input("Enter the common difference of the Arithmetic Progression: "))
    n = int(input("Enter the number of terms you want to calculate n-th term and sum of the frist n terms: "))
    t_n = a + (n-1)*d
    s_n = (n/2)*(2*a + (n-1)*d)
    return (f"N th term of the value is {t_n} and Sum of the first n th terms is {s_n}")
```

- Find the largest and smallest numbers in a list using a while loop or a for loop `(don't use max() or min())`.

```python
def numpositionfinder(number_list):
    smallest = number_list[0]
    largest = number_list[0]
    for num in number_list:
        if num > largest:
            largest = num
        if num < smallest:
            smallest = num
    return f"Smallest number: {smallest} and Largest number: {largest}"
```

- Use a while/for loop to calculate the factorial of a given number $n$. (e.g., $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$).

`For Loop`

```python
def factorialcalculator():
    fact = int(input("Plese a factorial number: "))
    factorial = 1
    for i in range(1,fact+1):
        factorial *= i
    return factorial
```

```python
factorialcalculator()
```

`While Loop`

```python
def factorialcalculator():
    fact = int(input("Plese a factorial number: "))
    factorial = 1
    i = 1
    while i < fact+1:
        factorial *= i
        i += 1
    return factorial
```

```python
factorialcalculator()
```

- $1 + 2 + 3 + 4 + 5 + \dots + N$. This is the foundational example showing how to calculate the sum of the first $N$ natural numbers.
- $Formula = \frac{n(n+1)}{2}$

```python
def naturalnumbersum(n):
    return (n*(n+1)/2)
```

```python
naturalnumbersum(100)
```

`For Loop`

```python
def naturalnumbersum(N):
    sum_total = 0
    for i in range(1,N+1):
        sum_total += i
    return sum_total
```

```python
naturalnumbersum(100)
```

`While Loop`

```python
def naturalnumbersum(N):
    sum_total = 0
    i = 1
    while i <= N:
        sum_total += i
        i += 1
    return sum_total
```

```python
naturalnumbersum(100)
```

- $ax^2 + bx + c = 0$ is a quadratic equation. Create an programe to determine the roots of this equation, which are given by the formulas:$$X1 = \frac{-b + \sqrt{b^2 - 4ac}}{2a}$$$$X2 = \frac{-b - \sqrt{b^2 - 4ac}}{2a}$$

The nature of the roots depends on the values of $(a, b, c)$. Specifically, $b^2 - 4ac$ is called the Discriminant (or $D$). Based on the sign of $D$, the nature of the roots (real, imaginary, or equal) can be determined.

**Input:** $(4x^{2}-5x-12=0)$

**Output:** $(X1 = -1.216, X2 = 2.466)$

```python
def quadratic():
    a = float(input("Please input value of a: "))
    b = float(input("Please input value of b: "))
    c = float(input("Please input value of c: "))

    if a == 0:
        if b == 0:
            print("No solution exists")
        else:
            x = round(-c / b, 3)
            print("Solution of this equation is:", x)
        return

    D = b**2 - 4*a*c

    if D == 0:
        x = round(-b / (2*a), 3)
        print("Solution of this equation is:", x)

    elif D > 0:
        x1 = round((-b + D**0.5) / (2*a), 3)
        x2 = round((-b - D**0.5) / (2*a), 3)
        print(f"Solutions of this equation are: {x1} and {x2}")

    else:
        print("This equation has imaginary solutions")
```

- $1 + 2 + 3 + 4 + 5 + \dots + N$. Write a programte to determine how many odds and even number in this series.

```python
def oddevenconter():
    a = float(input("Enter first term of the Arithmetic Progression: "))
    d = float(input("Enter the common difference of the Arithmetic Progression: "))
    n = int(input("Enter the last number of the Arithmetic Progression: "))
    even = 0
    odd = 0
    i = a
    while i <= n :
        if i % 2 == 0:
            even += 1
        else:
            odd += 1
        i += d
    print(f"Total Number of Even number is {even} and Odd is {odd}")
```
