### Calulation sign and usees:
<ul>
    <li>"+"  Add two or multiple values</li>
    <li>"-"  Substract two value</li>
    <li>"*"  Multiplication two or multiple values</li>
    <li>"/"  Division two values</li>
    <li>"%"  Modulo (5%2 = 1); Which gives reminder values</li>
    <li>"**" Exponential or Power</li>
    <li>">"  Greater Than (6>5)</li>
    <li>">=" Greater or equal</li>
    <li>"<"  Less Than (5<6)</li>
    <li>"<=" Less or equal</li>
    <li>"==" Equal to (5==5)</li>
    <li>"!=" Not equal to (5 != 6)</li>
</ul>
